```python
# ==============================
# 📦 IMPORTS (EDA)
# ==============================
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# ==============================
# 📂 LOAD CLEAN DATA
# ==============================
df = pd.read_csv("prepared.csv", parse_dates=["datetime"])

print("✅ Data loaded for EDA")
display(df.head())
```

##DISTRIBUTIONS OF ALL FEATURES

```python
# ==============================
# 📊 FEATURE DISTRIBUTIONS
# ==============================

features = [
    "temp", "dew_point", "wet_bulb",
    "rel_humidity", "spec_humidity",
    "pressure", "wind_speed"
]

for col in features:
    plt.figure()
    sns.histplot(df[col], bins=50)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

##STORM vs NON-STORM

```python
# ==============================
# 📊 STORM vs NON-STORM COMPARISON
# ==============================

for col in features:
    plt.figure()
    sns.boxplot(x="storm", y=col, data=df)
    plt.title(f"{col} vs Storm")
    plt.show()
```

##MONTHLY STORM TREND

```python
# ==============================
# 📅 MONTHLY STORM ANALYSIS
# ==============================

monthly_storms = df.groupby("month")["storm"].mean()

plt.figure()
monthly_storms.plot(kind='bar')

plt.title("Average Storm Occurrence by Month")
plt.xlabel("Month")
plt.ylabel("Storm Probability")
plt.show()
```

##TIME SERIES

```python
# ==============================
# 📈 TIME SERIES (PRECIPITATION)
# ==============================

plt.figure(figsize=(15,5))

plt.plot(df["datetime"], df["precipitation"])

plt.title("Precipitation Over Time")
plt.xlabel("Time")
plt.ylabel("Precipitation")
plt.show()
```

##STORM FREQUENCY OVER TIME

```python
# ==============================
# 📈 STORM EVENTS OVER TIME
# ==============================

plt.figure(figsize=(15,5))

plt.plot(df["datetime"], df["storm"])

plt.title("Storm Events Over Time")
plt.xlabel("Time")
plt.ylabel("Storm (0/1)")
plt.show()
```
